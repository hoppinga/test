```python
# Databricks Sample Notebook

# Load sample data
df = spark.read.format("csv").option("header", "true").load("/databricks-datasets/iris.csv")

# Display the dataframe
display(df)

# Select specific columns
df_selected = df.select("sepal_length", "sepal_width", "species")
display(df_selected)

# Group by species and calculate average sepal length
df_grouped = df.groupBy("species").agg({"sepal_length": "avg"})
display(df_grouped)

# Filter rows where sepal_length > 5.0
df_filtered = df.filter(df.sepal_length > 5.0)
display(df_filtered)
```
